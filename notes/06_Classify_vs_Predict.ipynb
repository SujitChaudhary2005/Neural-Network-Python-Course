## Classify vs. Predict

### **Understanding Predictors**

In the previous lesson, we built a **predictor**. A predictor takes an **input value** and produces a **numerical output**. For example, when converting kilometers to miles, the model used a linear function with an adjustable parameter to predict the correct number of miles. The prediction improved by **adjusting the internal parameter** based on the error between the predicted value and the known correct value.

This type of model focuses on predicting *how much* of something there is.

---

### **Introducing Classification**

Now we explore a different task: **classification**. Instead of predicting a numerical value, a classifier decides **which group something belongs to**.

Consider a dataset showing the **widths and lengths of garden bugs**.

![Dataset to be classified](../images/Dataset%20to%20be%20classified.png)

From the graph, two distinct groups are visible:
- **Caterpillars** are long and thin
- **Ladybugs** are short and wide

The goal is not to predict a number, but to **separate these two groups correctly**.

---

### **Using a Linear Function for Classification**

Earlier, we used a linear function to convert one number into another. A key property of a linear function is that it forms a **straight line when plotted**. The slope of this line is controlled by an adjustable parameter.

Instead of using the line to calculate values, we can use it as a **dividing boundary**.

![Prediction using a separating line](../images/Prediction%20using%20a%20separating%20line.png)

If a straight line can separate caterpillars from ladybugs, then it can be used to **classify unknown bugs** based on their measurements.

However, the first line shown does not separate the groups correctly. Many caterpillars appear on the same side of the line as ladybugs.

---

### **Adjusting the Line**

By changing the **slope of the line**, we change how the data is divided.

![Adjusting the slope](../images/Adjusting%20the%20slope.png)

Some adjustments make the separation worse. This demonstrates that **not all parameter values produce useful classifiers**.

After further adjustment, a much better line appears.

![Bug classifier](../images/Bug%20classifier.png)

This line cleanly separates caterpillars from ladybugs. Once this happens, the line becomes a **working classifier**.

---

### **Classifying Unknown Data**

Once a classifier is trained, it can be used on **new, unseen data**.

![Classifying an unknown bug](../images/Classifying%20an%20unknown%20bug.png)

If a new bugâ€™s measurements place it above the line, it can be classified as a caterpillar. If it falls below, it is classified as a ladybug. This assumes only the two known bug types exist, which is acceptable for illustrating the idea.

---

### **Why This Matters**

This lesson shows that the **same linear function** used for prediction can also be used for **classification**, depending on how it is applied. The critical challenge is **finding the right slope** so the line separates data correctly.

How we adjust that slope and improve poor separations is the central idea behind **how neural networks learn**.
